```python
def NOD (a, b):
    while (a & b):
        if (a >= b):
            a %= b
        else:
            b %= a
            
    return a | b
```

```python
#функция, котоая проверяет число на простоту
def is_prime(n):
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
#функция, которая возвращает простое рандомное число длины,
#введенной с клавиатуры и больше
import random

def looking_for_simple():
    #n = int(input())
    n = random.randint(0, 9)

    num = 1
    for i in range(1, n+1):
        num *= 10
        num += random.randint(0, 9)

    while not is_prime(num):
        num += 1

    return num
```

```python
#функция, которая ищет рандомное число из промежутка
# (1, fi-1), которое является взаимнопростым с fi
def looking_for_e(fi):
    e = random.randint(0, fi)

    a = fi
    b = e
    while (NOD(a, b) != 1) and (not is_prime(e)):
        e += 1
        a = fi
        b = e

    return e
```

```python
#функция, которая ищет обратный элемент в кольце остатков fi
def inverseE(fi, e):
    r = [fi, e]
    v = [0, 1]
    r.append(0)
    v.append(0)

    while r[2] != 1:
        q = r[0] // r[1]
        r[2] = r[0] % r[1]
        v[2] = v[0] - q*v[1]
        v[0] = v[1]
        v[1] = v[2]
        r[0] = r[1]
        r[1] = r[2]

    while v[2] < 0:
        v[2] += fi

    return v[2]
```

```python
#перевод из десятичной системы изсчисления в двоичную
def ten2two(num):
    bin = 0
    k = 1

    while num:
        bin += (num % 2) * k
        k *= 10
        num //= 2

    return bin
```
